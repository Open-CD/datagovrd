```python
def eliminar_columnas_no_nombradas(df):
    # Encontrar la primera fila que contiene datos
    primera_fila = df.apply(lambda x: x.notnull().sum(), axis=1).idxmax()
    
    # Retornar el DataFrame con las filas a partir de la primera fila
    df = df.iloc[primera_fila:]
    # Establecer la primera fila como los encabezados
    df.columns = df.iloc[0]
    # Eliminar la primera fila (ahora encabezados)
    df = df.iloc[1:]
    # Restablecer el índice
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
eliminar_columnas_no_nombradas(df)
```
